```python
import re
with open('row.txt','r', encoding='utf-8') as file:
    for line in file:
        a = r'^ab*$'
        b = line.strip()
        if re.fullmatch(a,b):
            print('True')
        else:
            print('False')
```

```python
import re 

def to_snake(s):
    return re.sub(r"(?<!^)([A-Z])", r"_\1", s).lower()

string = input()

print(to_snake(string))
```

```python
import re
def s(a):
    a = list(a)
    for i in range(len(a)-1):
        if a[i] == '_':
            a[i+1] = a[i+1].upper()
    res = ''.join(b for b in a if b!="_")
    return res
a = input()
print(s(a))
```
